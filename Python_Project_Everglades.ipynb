```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
file_path = "LT_CCD_Chambers.csv"
df = pd.read_csv(file_path)

# Group by 'SITENAME' and 'Year' and calculate the average %Carbon
grouped_df = df.groupby(['SITENAME', 'Year'])['%Carbon'].mean().reset_index()

# Get unique site names
site_names = grouped_df['SITENAME'].unique()

# Set the number of rows and columns in the grid
num_rows = 6
num_cols = 3

# Calculate the total number of subplots needed
total_subplots = len(site_names)

# Create subplots with the specified grid arrangement and larger figure size
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 20))

# Iterate over unique sites and plot the average %Carbon over time for each site
for i in range(total_subplots):
    site = site_names[i]
    site_data = grouped_df[grouped_df['SITENAME'] == site]

    # Set specific colors for certain sites
    if site in ["SRS3", "TS/Ph11", "TS/Ph6b", "TS/Ph7b", "TS/Ph9"]:
        axes[i // num_cols, i % num_cols].plot(site_data['Year'], site_data['%Carbon'], marker='o', label=site, color='blue')
    else:
        axes[i // num_cols, i % num_cols].plot(site_data['Year'], site_data['%Carbon'], marker='o', label=site, color='green')

    axes[i // num_cols, i % num_cols].axvline(2017.25, color='red', linestyle='--', label='Year 2017 Marker')

    axes[i // num_cols, i % num_cols].set_title(f'Average %Carbon - {site}')
    axes[i // num_cols, i % num_cols].set_xlabel('Year')
    axes[i // num_cols, i % num_cols].set_ylabel('Average %Carbon')
    axes[i // num_cols, i % num_cols].legend()
    axes[i // num_cols, i % num_cols].grid(True)

# Remove any empty subplots
for i in range(total_subplots, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Adjust layout for better spacing
plt.tight_layout()

# Save the plot as a PDF file (you can change the format if needed)
plt.savefig('combined_graphs_grid_6x3.pdf', bbox_inches='tight')

# Show the combined plot
plt.show()
```

```python
#perform t-test for %Carbon
carbon_before = before_Irma['%Carbon']
carbon_after = after_Irma['%Carbon']
t_stat_carbon, p_value_carbon = ttest_ind(carbon_before, carbon_after)

#perform t-test for %InorganicC
inorganicC_before = before_Irma['%InorganicC']
inorganicC_after = after_Irma['%InorganicC']
t_stat_inorganicC, p_value_inorganicC = ttest_ind(inorganicC_before, inorganicC_after)

#perform t-test for %OrganicC
organicC_before = before_Irma['%OrganicC']
organicC_after = after_Irma['%OrganicC']
t_stat_organicC, p_value_organicC = ttest_ind(organicC_before, organicC_after)

#print the results
print('T-Test results for %Carbon: t-statistic=', t_stat_carbon, ", p-value =", p_value_carbon)
print('T-Test results for %InorganicC: t-statistic=', t_stat_inorganicC, ", p-value =", p_value_inorganicC)
print('T-Test results for %OrganicC: t-statistic=', t_stat_organicC, ", p-value =", p_value_organicC)
```

```python
from scipy.stats import ttest_ind

# Filter data before and after 2017
before_Irma = df[df['Year'] < 2017.0]
after_Irma = df[df['Year'] >= 2017.5]

# List to store t-test results for each site
ttest_results = []

# Perform t-test for %Carbon for each site
for site in df['SITENAME'].unique():
    carbon_before = before_Irma[before_Irma['SITENAME'] == site]['%Carbon']
    carbon_after = after_Irma[after_Irma['SITENAME'] == site]['%Carbon']
    
    # Perform t-test
    t_stat, p_value = ttest_ind(carbon_before, carbon_after, equal_var=False)  # Assuming unequal variance
    
    # Store t-test results for each site
    ttest_results.append({'Site': site, 'T-Statistic': t_stat, 'P-Value': p_value})

# Display t-test results for each site
for result in ttest_results:
    print(f"Site: {result['Site']}, T-Statistic: {result['T-Statistic']}, P-Value: {result['P-Value']}")
```

```python
# Bonferroni correction, applited by dividing original significance level by the number of tests 
# Compare each p-value to the Bonferroni-corrected significance level and determine p-values that are still statistically significant:

ttest_results

p_values = []
for result in ttest_results:
    p_value = result['P-Value']
    p_values.append(p_value)
    

alpha = 0.05

alpha_bonferroni = alpha / len(p_values)
significant_tests = []      
for p_value in p_values:
    if p_value <= alpha_bonferroni:
        significant_tests.append(p_value)
        
        
print(significant_tests) 
print("The Bonferroni-corrected significance level is:", alpha_bonferroni)

# associtating the p-values that are still significant back to the site name so it prints out nicely :) 
p_values = []
for result in ttest_results:
    p_value = result['P-Value']
    site = result['Site']
    p_values.append((p_value, site))

matching_tuples = [(p, s) for p, s in p_values if p in significant_tests]

if matching_tuples:
    for t in matching_tuples:
        print(f"P-Value: {t[0]}, Site: {t[1]}")
else:
    print("No matching tuples found for the specified p-values.") 
```

```python
# repeated measures ANOVA 
import pandas as pd
from statsmodels.stats.anova import AnovaRM
df

# Assuming your 'Year' column is in string format and your DataFrame is named 'df'
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

#changing dataframe to add new column categorizing for before and after hurricane 

def categorize_beforeafter_hurricane(year):
    if year == 2017.0:
        return 'Before'
    if year == 2017.5:
        return 'After'

df['BeforeAfter'] = df['Year'].apply(categorize_beforeafter_hurricane)

df['BeforeAfter'] = df['Year'].apply(categorize_beforeafter_hurricane)

# aggregate data
aggregated_data = df.groupby(['SITENAME', 'BeforeAfter'])['%Carbon'].mean().reset_index()


# removing site TS/Ph8 because they only contain data in 2017 from before the Hurricane, but not after 
aggregated_data = aggregated_data.drop(30)

# perform repeated measures ANOVA 
anova = AnovaRM(aggregated_data, depvar='%Carbon', subject='SITENAME', within=['BeforeAfter'])
results = anova.fit()

# view the results, the Pr > F (p-value) is the p-value associated with the F-statistic (to be considered significant, should be below 0.05) 
print(results)

```
